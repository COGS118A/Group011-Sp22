```python
import pandas as pd
from numpy import asarray
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('heart_2020_cleaned.csv')
```

```python
df.head()
```

```python
encoder = OneHotEncoder(handle_unknown='ignore')
```

```python
heart_disease_df= pd.DataFrame(encoder.fit_transform(df[['HeartDisease']]).toarray())
heart_disease_name = ["Heart disease(No)", "Heart disease(Yes)"]
heart_disease_df.columns = heart_disease_name
print(heart_disease_df.shape)
heart_disease_df.head()
```

```python
smoking_df= pd.DataFrame(encoder.fit_transform(df[['Smoking']]).toarray())
smoking_name = ["Smoke(No)", "Somke(Yes)"]
smoking_df.columns = smoking_name
print(smoking_df.shape)
smoking_df.head()
```

```python
alcohol_drinking_df= pd.DataFrame(encoder.fit_transform(df[['AlcoholDrinking']]).toarray())
alcohol_drinking_name = ["AlcoholDrinking(No)", "AlcoholDrinking(Yes)"]
alcohol_drinking_df.columns = alcohol_drinking_name
print(alcohol_drinking_df.shape)
alcohol_drinking_df.head()
```

```python
stroke_df = pd.DataFrame(encoder.fit_transform(df[['Stroke']]).toarray())
stroke_name = ["Stroke(No)", "Stroke(Yes)"]
stroke_df.columns = stroke_name
print(stroke_df.shape)
stroke_df.head()
```

```python
diff_walking_df = pd.DataFrame(encoder.fit_transform(df[['DiffWalking']]).toarray())
diff_walking_name = ["DiffWalking(No)", "DiffWalking(Yes)"]
diff_walking_df.columns = diff_walking_name
print(diff_walking_df.shape)
diff_walking_df.head()
```

```python
sex_df = pd.DataFrame(encoder.fit_transform(df[['Sex']]).toarray())
sex_name = ["Female", "Male"]
sex_df.columns = sex_name
print(sex_df.shape)
sex_df.head()
```

```python
df["AgeCategory"].unique()
```

```python
df['AgeCategory'] = df['AgeCategory'].replace("80 or older", "80")
df['AgeCategory'] = df['AgeCategory'].str.rsplit("-")
df['AgeCategory'] = df['AgeCategory'].str[0]
df['AgeCategory'] = df['AgeCategory'].astype('float64')
df.head()
```

```python
def one_hot_encode(x:np.ndarray):
    uni = np.unique(x)
    print(uni)
    result = np.zeros((len(x), len(uni)))
    for i in range(len(x)):
        for j in range(len(uni)):
            if x[i] == uni[j]:
                result[i][j] = 1;
                
    return result
```

```python
race_arr = pd.DataFrame(df['Race']).to_numpy()
one_hot_race = one_hot_encode(race_arr)
race_df = pd.DataFrame(one_hot_race, columns = ['American Indian/Alaskan Native', 'Asian', 'Black', 'Hispanic', 'Other',
 'White'])
race_df.head()
```

```python
diabetic_arr = pd.DataFrame(df['Diabetic']).to_numpy()
one_hot_diabetic = one_hot_encode(diabetic_arr)
diabetic_df = pd.DataFrame(one_hot_diabetic, columns = ['Diabetic(No)', 'Diabetic(No, borderline diabetes)', 'Diabetic(Yes)', 'Diabetic(Yes, during pregnancy)'])
diabetic_df.head()
```

```python
physical_activity_df= pd.DataFrame(encoder.fit_transform(df[['PhysicalActivity']]).toarray())
physical_activity_name = ["Physical Activity(No)", "Physical Activity(Yes)"]
physical_activity_df.columns = physical_activity_name
print(physical_activity_df.shape)
physical_activity_df.head()
```

```python
gen_health_arr = pd.DataFrame(df['GenHealth']).to_numpy()
one_hot_gen_health = one_hot_encode(gen_health_arr)
gen_health_df = pd.DataFrame(one_hot_gen_health, columns = ['GenHealth(Excellent)', 'GenHealth(Fair)', 
                                                            'GenHealth(Good)', 'GenHealth(Poor)', 'GenHealth(Very good)'])
```

```python
asthma_df= pd.DataFrame(encoder.fit_transform(df[['Asthma']]).toarray())
asthma_name = ["Asthma(No)", "Asthma(Yes)"]
asthma_df.columns = asthma_name
print(asthma_df.shape)
asthma_df.head()
```

```python
kidney_disease_df= pd.DataFrame(encoder.fit_transform(df[['KidneyDisease']]).toarray())
kidney_disease_name = ["KidneyDisease(No)", "KidneyDisease(Yes)"]
kidney_disease_df.columns = kidney_disease_name
print(kidney_disease_df.shape)
kidney_disease_df.head()
```

```python
skin_cancer_df= pd.DataFrame(encoder.fit_transform(df[['SkinCancer']]).toarray())
skin_cancer_name = ["SkinCancer(No)", "SkinCancer(Yes)"]
skin_cancer_df.columns = skin_cancer_name
print(skin_cancer_df.shape)
skin_cancer_df.head()
```
